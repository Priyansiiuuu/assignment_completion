```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("assignment_dataset.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['Due_Date'] = pd.to_datetime(df['Due_Date'])
df['Submission_Date'] = pd.to_datetime(df['Submission_Date'])
#pd.to_datetime() converts a date column from string format to datetime format,
#allowing date comparison, subtraction,
#and time-based analysis such as finding late submission days.
```

```python
df
```

```python
df['Late_Days'] = (df['Submission_Date'] - df['Due_Date']).dt.days

df['Late_Days'] = df['Late_Days'].apply(lambda x: x if x > 0 else 0)
```

```python
df
```

```python
df.describe()
```

```python
df['Status'].value_counts()
```

```python
sns.countplot(x='Status', data=df)
plt.title('Assignment Submission Status')
plt.show()
```

```python
sns.histplot(df['Late_Days'], bins=5, kde=True)
plt.title('Late Submission Trend')
plt.show()
```

### Conclusion
- Most students submit assignments on time.
- Late submissions show a trend of 24 days delay.
- Visualization helps identify submission behavior.
- This analysis can help instructors improve deadline policies.
